```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


sample_data=pd.read_csv("dataset.csv",nrows=1000)
missing_values = sample_data.isnull()
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.boxplot(sample_data['apartment_count'], vert=False)
plt.title('Ящик с усами')

plt.subplot(1, 2, 2)
plt.hist(np.log1p(sample_data['apartment_count']), bins=20, edgecolor='black')
plt.title('Гистограмма (логарифмическая шкала)')

plt.show()

sample_data['apartment_count'].fillna(sample_data['apartment_count'].median(), inplace=True)
sample_data['apartment_count'] = np.clip(sample_data['apartment_count'], 1, 20) 


room_counts = sample_data['room_count'].value_counts()

pivot_table = pd.pivot_table(sample_data, values='apartment_count', index='district', columns='room_count', aggfunc='count')

pivot_table.to_csv("surname.csv")
```
