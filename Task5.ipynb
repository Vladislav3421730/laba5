```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D





x=np.arange(0,41,2)
y=np.arange(0,61,3)
z=np.power(x,0.25)+np.power(y,0.25)
fig=plt.figure(figsize=(10, 6))
ax=fig.add_subplot(111,projection='3d')
ax.plot(x,y,z,label='parametric curve')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title("График функции z=x^0.25+y^0.25")

z=x**2-y**2
fig=plt.figure(figsize=(10, 6))
ax1=fig.add_subplot(111,projection='3d')
ax1.plot(x,y,z,label='parametric curve')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
ax1.set_title("График функции z=x^2-y^2")

z=2*x+3*y
fig=plt.figure(figsize=(10, 6))
ax2=fig.add_subplot(111,projection='3d')
ax2.plot(x,y,z,label='parametric curve')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')
ax2.set_title("График функции z=2x+3y")

z=x**2+y**2
fig=plt.figure(figsize=(10, 6))
ax3=fig.add_subplot(111,projection='3d')
ax3.plot(x,y,z,label='parametric curve')
ax3.set_xlabel('X')
ax3.set_ylabel('Y')
ax3.set_zlabel('Z')
ax3.set_title("График функции z=x^2+y^2")

z=2+2*x+2*y-x**2-y**2
fig=plt.figure(figsize=(10, 6))
ax4=fig.add_subplot(111,projection='3d')
ax4.plot(x,y,z,label='parametric curve')
ax4.set_xlabel('X')
ax4.set_ylabel('Y')
ax4.set_zlabel('Z')
ax4.set_title("График функции z=2+2x+2y-x^2-y^2")





```
