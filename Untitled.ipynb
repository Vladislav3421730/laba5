```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
x=np.linspace(-5,5,101)
y=x ** 2
y2=x ** 3
plt.plot(x,y)
plt.show()
plt.plot(x,y)
plt.plot(x,y2)
plt.title("Графики функций")
plt.xlabel("переменная x")
plt.ylabel("Функция f(x)")
plt.legend(labels=["x^2","x^3"])
plt.axis([-5,5,-8,8])
plt.show()
fig,ax=plt.subplots(nrows=1,ncols=2)
fig.set_size_inches(8.5,3.5)
fig.subplots_adjust(wspace=0.3)
ax[0].plot(x,y)
ax[1].plot(x,y2)
y3=np.sin(x)
y4=np.cos(x)

fig , ax =plt.subplots(nrows=2,ncols=2)
ax1,ax2,ax3,ax4 = ax.flatten()
fig.set_size_inches(8,6)
fig.subplots_adjust(wspace=0.3,hspace=0.3)
ax1.plot(x,y)
ax1.set_title("График x^2")
ax1.set_ylim([0,25])
ax1.set_xlim([-10,10])
ax2.plot(x,y2)
ax2.set_title("График x^3")
ax3.plot(x,y3)
ax3.set_title("График sin(x)")
ax4.plot(x,y4)
ax4.set_title("График cos(x)")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
x=np.linspace(-np.pi,np.pi,50)
y=x
z=np.cos(x)
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.plot(x,y,z,label='parametric curve')

u,v=np.mgrid[0:2*np.pi:20j,0:np.pi:10j]
x=np.cos(u)* np.sin(v)
y=np.sin(u)* np.sin(v)
z=np.cos(v)
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.plot_wireframe(x,y,z,cmap='inferno')
```
